```python
df1['Category'].value_counts()
```

```python
df1['Category'].replace('^\"',np.nan)
```

```python
if df1["Category"].dtype == 'object' and df1["Category"].str.startswith('"').any() :
    # remove the double quotes at the beginning of the string using str.replace() function
    df1["Category"] = df1["Category"].str.replace('^\"', '', regex=True)
```

```python
if df1["Category"].dtype == 'object' and df1["Category"].str.endswith('"').any() :
    # remove the double quotes at the end of the string using str.replace() function
    df1["Category"] = df1["Category"].str.replace('"$', '', regex=True)
```

## Figure out the Top 5 performing categories

```python
 Merge2['Category'].value_counts()
```

```python
As we can see above that the feature contains data which is repetative to overcome this we are we are converting all of them into lower case .
```

```python
Merge2['Category']=Merge2["Category"].str.lower()
```

```python
category_scores = Merge2.groupby('Category')['Score'].sum()
category_scores.sort_values(ascending=False)
```

```python
Merge2.to_csv("data_cleaning_set.csv")
```

# Visualizations to understand Data Better

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Categories = Merge2['Category'].unique()
Categories = sorted(Categories)
```

```python
category_scores
```

```python
sns.barplot(x=category_scores,y=Categories)
plt.xlabel("Score")
plt.ylabel("Category")

plt.show()
```

```python
plt.pie(category_scores,labels=Categories,autopct='%1.1f%%') # it is used to show values as percentages
plt.show()
```
